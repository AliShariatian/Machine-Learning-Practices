### <a id='toc1_5_3_'></a>[Principal Component Analysis (PCA)](#toc0_)

Variance = Importance = Information

```python
np.random.seed(1)
X = (np.random.random(size=(2, 2)) @ np.random.normal(size=(2, 200))).T

plt.plot(X[:, 0], X[:, 1], 'o')
plt.axis('equal')
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # get 2 important components from all components
pca.fit(X)

print('eigen value (informations in features):')
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
print('97.6% information in feature 0 \n02.3% information in feature 1')
print()

print('eigen vector:')
print(pca.components_)
```

```python
plt.bar(range(1, len(pca.explained_variance_ratio_)+1),
        pca.explained_variance_ratio_,
        alpha=0.5,
        align='center')

plt.step(range(1, len(pca.explained_variance_ratio_)+1),
         np.cumsum(pca.explained_variance_ratio_),
         where='mid')


plt.xticks([1, 2])
plt.yticks(np.arange(0, 1.01, 0.05))
plt.show()
```

Other way:

```python
clf = PCA(0.95)  # keep 95% of variance (information)
X_trans = clf.fit_transform(X)

print(X.shape)
print(X_trans.shape)
```

```python
X_new = clf.inverse_transform(X_trans)
```

```python
print(X_new.shape)
```

```python
print(X.mean(axis=0))
print(X_trans.mean(axis=0))
print(X_new.mean(axis=0))
```

```python
plt.plot(X[:, 0], X[:, 1], 'o', alpha=0.2)
plt.plot(X_new[:, 0], X_new[:, 1], 'o', alpha=0.8)

plt.axis('equal')
plt.show()
```

Elbow Method

```python
distoration = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, random_state=0, n_init='auto')
    km.fit(X)
    distoration.append(km.inertia_)
```

```python
import keras
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
image_size = 32
num_chanels = 3
num_classes = 10
num_features = num_chanels * image_size**2

num_train = 49_000
```

```python
print(num_features)
```

```python
data = cifar10.load_data()
```

```python
print(type(data))
print(np.shape(data))
print(np.ndim(data))
```

```python
(X_train, y_train), (X_test, y_test) = data


print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

Data visualization

```python
class_names = ['airplane', 'automobile', 'bird', 'cat',
               'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

samples_per_class = 5
```

```python
plt.figure(figsize=(16, 8))

for cls, name in enumerate(class_names):
    idxs = np.flatnonzero(y_train == cls)
    idxs = np.random.choice(idxs, samples_per_class, replace=False)

    for i, idx in enumerate(idxs):
        plt.subplot(samples_per_class, num_classes, i * num_classes + cls + 1)
        plt.imshow(X_train[idx], interpolation='spline16')
        plt.axis('off')
        if i == 0:
            plt.title(class_names[cls])
```

Data Preprocessong

```python
print('X_train shape Before reshape', X_train.shape)
print('X_test shape Before reshape', X_test.shape)
print()

# Convert 4D arrays to 2D arrays
X_train = X_train.reshape([-1, num_features])  # ? * 3072 Matrixs
X_test = X_test.reshape([-1, num_features])

print('X_train shape After reshape', X_train.shape)
print('X_test shape After reshape', X_test.shape)
```

```python
# Type changing
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
mu = np.mean(X_train, axis=0)

X_train = X_train - mu
X_train = X_train / 255.0

X_test = X_test - mu
X_test = X_test / 255.0
```

```python
print(y_train.shape)
print(y_test.shape)

# One-hot encodding
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print(y_train.shape)
print(y_test.shape)
```

```python
[print(y_train[i]) for i in range(4)]
```

Linear Classifier

```python
model = Sequential()  # A sequence of layers
model.add(Dense(num_classes, input_shape=(num_features,),
                activation='softmax'))
```
